```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans

# Load the CSV files
df1 = pd.read_csv("Term Test-1-result.csv")
df2 = pd.read_csv("Term Test-2-result.csv")
```

```python
df_merged = pd.merge(df1, df2, on=["Registration Number", "Name"])
print(df_merged)
```

```python
df_merged.rename(columns={"TT 1 Marks ": "TT-1 Marks", "TT 2 Marks": "TT-2 Marks"}, inplace=True)
print(df_merged.columns)
```

```python
df_merged["Best Marks"] = df_merged[["TT-1 Marks", "TT-2 Marks"]].max(axis=1)
df_merged["Average Marks"] = df_merged[["TT-1 Marks", "TT-2 Marks"]].mean(axis=1)
print(df_merged)
```

```python
df_final = df_merged.drop(columns=["TT-1 Marks", "TT-2 Marks"],inplace=True)
print(df_final)
```

```python
df_attendence = pd.read_csv("attendence-term-final.csv")
df_final =pd.merge(df_merged,df_attendence, on =['Registration Number','Name'])
df_final['Total Marks']= df_final ['Final Marks'] + df_final['Average Marks'] + df_final ['Attendence Marks']
print(df_final)
```

```python
df_final.to_csv("final result.csv", index= False)
```

```python
import pandas as pd
final= pd.read_csv("final result.csv")
final
```

```python
x=final.index.tolist()
y= final["Total Marks"].tolist()
x,y
data= list(zip(x,y))
data
```

```python
import matplotlib.pyplot as plt
plt.scatter(x,y)
plt.title('Final Marks of Students')
plt.xlabel('Student Index')
plt.ylabel('Final Marks')
plt.show()
```

```python
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt

inertias = []
for i in range(1, 11):
    kmeans = KMeans(n_clusters=i)
    kmeans.fit(data)
    inertias.append(kmeans.inertia_)

plt.plot(range(1, 11), inertias, marker='o')
plt.title('Elbow Method')
plt.xlabel('Number of Clusters')
plt.ylabel('Inertia')
plt.show()
```

```python
kmeans = KMeans(n_clusters=3, n_init=10)
kmeans.fit(data)
labels = kmeans.labels_

plt.scatter(x, y, c=labels, cmap='viridis')

for cluster in range(3):
    plt.scatter([], [], color=plt.cm.viridis(cluster / 2), label=f'Cluster {cluster}')

plt.title('K-Means Clustering of Final Marks')
plt.xlabel('Student Index')
plt.ylabel('Final Marks')

plt.legend(title='Clusters', loc='center left', bbox_to_anchor=(1, 0.5))

plt.show()
```
